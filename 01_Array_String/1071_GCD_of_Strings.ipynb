# 1071. Merge_Strings_Alternately

**Problem Link**: [LeetCode Link](https://leetcode.cn/problems/greatest-common-divisor-of-strings/description/?envType=study-plan-v2&envId=leetcode-75)

## Problem Description



```python
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        len1, len2 = len(str1), len(str2)
        i,j = 0,0

        #Step 2: Find GCD of the lengths
        # def gcd(a, b):
        #     while b:
        #         a, b = b, a % b
        #     return a

        # gcd_length = gcd(len1, len2)
        
        gcd_length = math.gcd(len1, len2)   #O(log(m+n))

        # Step 3: Get the candidate string
        candidate = str1[:gcd_length] 
        
        #O(gcd_length) gcd_length minm,n


        # Step 4: Verify the candidate
        # Check if repeating the candidate forms str1
        if candidate * (len1 // gcd_length) != str1:
            return ""
        # O(m)
        
        # Check if repeating the candidate forms str2
        if candidate * (len2 // gcd_length) != str2:
            return ""
        #O(n)

        
        # Step 5: If both checks pass, this is the answer
        return candidate

        
        
```


## Complexity Analysis
 str1 m,str2 n
- **Time Complexity**: $O(m+n)$

- **Space Complexity**: $O(m+n)$
