## Complexity Analysis

- **Time Complexity**: $O(n)$

- **Space Complexity**: $O(1)$

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        length = len(nums)
        output = [1] * length
        
        left = 1
        for i in range(length):
            output[i] = left
            left = left * nums[i] 

        right = 1
        
        for i in range(length - 1, -1, -1):
            output[i] = output[i] * right
            right = right * nums[i] 
            
        return output
```
