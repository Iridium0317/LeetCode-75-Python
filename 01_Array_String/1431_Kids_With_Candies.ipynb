```python
# [1431]. [Kids With the Greatest Number of Candies]

[LeetCode Link](https://leetcode.cn/problems/kids-with-the-greatest-number-of-candies?envType=study-plan-v2&envId=leetcode-75)

## Problem Description


## Complexity Analysis

- **Time Complexity**: $O(m+n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...
```

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max_candies = max(candies)
        result = []
        for candy in candies:
            if candy + extraCandies >= max_candies:
                result.append(True)

            else:
                result.append(False)
        return result       

        
```
