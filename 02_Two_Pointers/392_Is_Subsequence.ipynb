```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        i, j = 0, 0
        n, m = len(s), len(t)
        
        # 
        while i < n and j < m:
            if s[i] == t[j]:
                i += 1
            j += 1
            
        #  i  n s 
        return i == n
```
