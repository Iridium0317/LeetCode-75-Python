## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        slow = 0
        
        for fast in range(len(nums)):
    
            if nums[fast] != 0:
                key = nums[fast]
                nums[fast] = nums[slow]
                nums[slow] = key
                %nums[slow], nums[fast] = nums[fast], nums[slow]
                slow+=1 
```
