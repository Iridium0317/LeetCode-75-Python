```python
class Solution:
    def maxOperations(self, nums: List[int], k: int) -> int:
        # 1. 
        nums.sort()

        count = 0
        left = 0
        right = len(nums)-1
        while left<right:
            current_sum = nums[left] + nums[right]

            if current_sum==k:
                count+=1
                left+=1
                right-=1
              
            
            elif current_sum < k:
                # 
                # 
                left += 1
                
            else: # current_sum > k
                # 
                # 
                right -= 1
        return count
        
```
