## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left = 0
        right = len(height) - 1 
        max_s = 0 
        
        while left < right:
            w = right - left  
            h = min(height[left], height[right])
            
            current_s = w * h
            max_s = max(max_s, current_s)
            
            #  
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
                
        return max_s  
```
