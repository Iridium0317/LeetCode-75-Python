## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def closeStrings(self, word1: str, word2: str) -> bool:
        if len(word1) != len(word2):
            return False

        from collections import Counter
        count1 = Counter(word1) # O(N)
        count2 = Counter(word2)
        set_key1 = set(count1) 
        set_key2 = set(count2)
        set_val1 = sorted(count1.values()) #O(klogk)
        set_val2 = sorted(count2.values())

        return  set_key1 == set_key2 and set_val1 == set_val2

        # space_complexity O(1)
        # time_complexity O(N) 
            
        
```
