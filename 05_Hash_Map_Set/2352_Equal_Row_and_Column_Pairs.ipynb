## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def equalPairs(self, grid: List[List[int]]) -> int:
        count = 0

        from collections import Counter
        
        row_counts = Counter()
        #æå–è¡Œ
        for row in grid:
            row_tuple = tuple(row)
            row_counts[row_tuple] += 1

        #æå–åˆ—    
        n = len(grid)
        col_counts = Counter()
        for j in range(n):
            col = []
            for i in range(n):
                col.append(grid[i][j])
            col_tuple = tuple(col)
            col_counts[col_tuple] += 1

        for tpl in row_counts:
            count += row_counts[tpl] * col_counts[tpl]
                
        return count
```
