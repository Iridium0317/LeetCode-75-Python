- **Space Complexity**: $O(n)$

```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        from collections import Counter
        counts = Counter(arr)
        occurrence_list = counts.values()
        set_count = set(occurrence_list)
        return len(set_count) == len(occurrence_list)
```
