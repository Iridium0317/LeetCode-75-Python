## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        set1 = set(nums1)
        set2 = set(nums2)

        diff1 = list(set1 - set2)
        diff2 = list(set2 - set1)
        return [diff1, diff2]
```
