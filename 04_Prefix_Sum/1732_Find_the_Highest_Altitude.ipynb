## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def largestAltitude(self, gain: List[int]) -> int:

        altitudes = [0]*(len(gain)+1)
        altitudes[0] = 0
        for i in range(len(gain)):
            altitudes[i+1] = altitudes[i]+gain[i]

        return max(altitudes)
```
