## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def maxVowels(self, s: str, k: int) -> int:
        vowels = "aeiou"
        count = 0
        for i in range(k):
            if s[i] in vowels:
                count+=1
        max_count = count

        if count == k:
            return k
    
        for i in range(1, len(s)-k+1):     
            if s[i-1] in vowels:
                count -= 1
                            
            if s[i+k-1] in vowels:
                count += 1
        
            max_count = max(max_count, count)
        
        if max_count == k:
            return k

        return max_count 


    
```
