- **Space Complexity**: $O(1)$

```python
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        current_sum = 0
        for i in range(k):
            current_sum += nums[i] 
        max_sum = current_sum
        
        for i in range(1, len(nums) - k + 1):
            
            current_sum = current_sum - nums[i-1] + nums[i+k-1]
            if current_sum > max_sum:
                max_sum = current_sum
                
        return max_sum / k
```
