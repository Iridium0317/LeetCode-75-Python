## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def longestOnes(self, nums: List[int], k: int) -> int:
        left = 0
        zeros_count = 0  
        max_len = 0

        for right in range(len(nums)):
 
            if nums[right] == 0:
                zeros_count += 1
            
            while zeros_count > k:
               
                if nums[left] == 0:
                    zeros_count -= 1              
                left += 1

            current_len = right - left + 1
            max_len = max(max_len, current_len)
            
        return max_len
        
```
