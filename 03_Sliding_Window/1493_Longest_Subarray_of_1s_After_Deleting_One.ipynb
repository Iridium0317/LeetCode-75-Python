## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        left = 0
        count = 0
        max_len = 0

        for right in range(len(nums)):
            if nums[right] == 0:
                count +=1

            while count >1:    
                
                if nums[left] == 0: #  0
                    count -= 1
                left += 1           # 

            max_len = max(right-left,max_len)    
            
        
        return max_len

        
```
