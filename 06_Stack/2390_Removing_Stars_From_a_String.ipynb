## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(n)$
  - *Reasoning*: list 


```python
class Solution:
    def removeStars(self, s: str) -> str:
        stack = [] 
        
        for char in s: # O(N)
            if char == '*':
                stack.pop()
            else:
                stack.append(char)
        
        return "".join(stack) # O(N)
```
