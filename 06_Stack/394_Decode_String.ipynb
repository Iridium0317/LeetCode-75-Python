## Complexity Analysis

- **Time Complexity**: $O(n)$
  - *Reasoning*: ...
- **Space Complexity**: $O(1)$
  - *Reasoning*: ...


```python
class Solution:
    def decodeString(self, s: str) -> str:
        stack = []       
        curr_num = 0     
        curr_str = ""    
        
        for char in s:
            #  10
            if char.isdigit():
                curr_num = curr_num * 10 + int(char)
            
            elif char == '[':
                # Tuple  append
                stack.append((curr_str, curr_num))
                curr_str = ""
                curr_num = 0

            elif char == ']':
                #Pythonunpacking
                prev_str, prev_num = stack.pop()
                curr_str = prev_str + (curr_str * prev_num)
            
            else:
                curr_str += char
                
        return curr_str

        
```
