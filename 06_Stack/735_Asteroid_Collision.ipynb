```python
class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        result = []
        
        for ast in asteroids:
            while result and result[-1] > 0 and ast < 0:
                
                if result[-1] < -ast: 
                    result.pop()
                    
                elif result[-1] == -ast:
                    result.pop()                   
                    break 
                
                else:
                    break
                
            # breakelse
            else:
                
                result.append(ast)
                
        return result
```
